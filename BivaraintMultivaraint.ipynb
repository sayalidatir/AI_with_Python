```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = tips['total_bill']  # x-coordinates
y = tips['tip']  # y-coordinates
z = tips['size']  # z-coordinates

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Three Variable Scatter Plot')

# Display the plot
plt.show()
```

# 2.bar plot:(categorical-NUmerical data)

```python
titanic
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic)

# Add labels and title
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Age Distribution by Passenger Class')

# Display the plot
plt.show()
```

```python
#boxplot
sns.boxplot(x='Gender', y='Age', data=titanic)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Passenger Class')

# Display the plot
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', hue='Gender', data=titanic)

# Add labels and title
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Age Distribution by Passenger Class and Gender')

# Display the plot
plt.show()
```
