```python
# standardization : It is one of the technique of feature scaling

import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
#split data into train data and split data
y=df['Purchased']
x=df.drop('Purchased',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_train.shape,x_test.shape
```

# standardScaler

```python
df
```

```python
#the standardization done before the EDA it is done in preprocessing satge
#by using satndardization we can scale the values of two columns

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(x_train)
#both training and testing data needs to be transform
#here we transform means convert column data into  another column(i.e age to 'age) formula =xi/x(mean)/std
x_train_scale=scaler.transform(x_train)
x_test_scale=scaler.transform(x_test)
```

```python
scaler.mean_#here the mean of two columns have more difference
```

```python
x_train
```

```python
x_train_scale
```

```python
X_train_scaled = pd.DataFrame(x_train_scale, columns=x_train.columns)
X_test_scaled = pd.DataFrame(x_test_scale, columns=x_test.columns)
```

```python
X_train_scaled
```

```python
x_train
```

```python
np.round(x_train.describe(), 1)#get mean of train x data i.e 0.0
```

```python
np.round(X_train_scaled.describe(), 1)#mean of scaled x_train data i.e also 0.0
```

# EFFECT OF SCALING:

```python
#we does not get effect on graph ,we get effect on the range of graph


fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

# COMPARISON OF DISTRIBUTION:

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
Why scaling is important:
    here  we use the scaling in logestic model  and calculate accuracy of it ,so it can give the more accuracy after scaling 
the value.
```

```python
from sklearn.linear_model import LogisticRegression

l_model=LogisticRegression()
l_model_scaled=LogisticRegression()

l_model.fit(x_train,y_train)
l_model_scaled.fit(x_train_scale,y_train)
```

```python
y_pred=l_model.predict(x_test)
```

```python
y_pred_scaled=l_model_scaled.predict(x_test_scale)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('accuracy of train data:',accuracy_score(y_test,y_pred))
print('accuracy of train data:',accuracy_score(y_test,y_pred_scaled))
```

```python
#if we use in decision tree it not affect on accuracy of data
```

```python
from sklearn.tree import DecisionTreeClassifier

d_tree=DecisionTreeClassifier()
dt_scaled=DecisionTreeClassifier()
d_tree.fit(x_train,y_train)
dt_scaled.fit(x_train_scale,y_train)
```

```python
y_pred=d_tree.predict(x_test)
```

```python
y_pred_scale=dt_scaled.predict(x_test_scale)
```

```python
print('accuracy:',accuracy_score(y_test,y_pred))
print('accuracy:',accuracy_score(y_test,y_pred_scale))
```
